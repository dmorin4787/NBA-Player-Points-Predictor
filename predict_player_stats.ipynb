```python
import pandas as pd

pd.set_option('display.max_columns', None)

df = pd.read_csv('player_box_scores.csv')
```

```python
def create_target(row):
    row = row.sort_values(by='Date')
    row['target'] = row['PTS'].shift(-1)
    return row

grouped_df = df.groupby('Player ID', group_keys=False).apply(create_target)
```

```python
grouped_df
```

```python
nulls = pd.isnull(df)
nulls = nulls.sum()
nulls[nulls > 0]
```

```python
filtered_df = grouped_df[grouped_df['Date'] >= '1996-09-01']
```

```python
nulls = pd.isnull(filtered_df)
nulls = nulls.sum()
nulls = nulls[nulls > 0]
```

```python
filtered_df[filtered_df['+/-'].isnull()]
```

```python
#Since +/- is important, we will impute the values with a Linear Regression model
from sklearn.linear_model import LinearRegression

# Select relevant features for prediction
features = ['Result', 'GS', 'MP', 'FG', 'FGA', '3P', '3PA', 'FT', 'FTA', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'GmSc']

# Separate rows with and without missing `+/-`
df_not_null = filtered_df[filtered_df['+/-'].notna()]
df_null = filtered_df[filtered_df['+/-'].isna()]

# Train a model
model = LinearRegression()
model.fit(df_not_null[features], df_not_null['+/-'])

# Predict missing values
filtered_df.loc[filtered_df['+/-'].isna(), '+/-'] = model.predict(df_null[features])
```

```python
filtered_df.loc[:, 'target'] = filtered_df['target'].fillna(-1)
```

```python
nulls = pd.isnull(filtered_df)
nulls = nulls.sum()
nulls = nulls[nulls > 0]
nulls
```

```python
valid_cols = filtered_df.columns[~filtered_df.columns.isin(nulls.index)]
```

```python
valid_cols
```

```python
predictor_df = filtered_df[valid_cols].copy()
predictor_df['Away?'] = predictor_df['Away?'].astype(int)
predictor_df['Season'] = pd.to_datetime(predictor_df['Date']).dt.year
predictor_df = predictor_df.drop('Unnamed: 0', axis=1)
```

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.feature_selection import SequentialFeatureSelector


model = LinearRegression()
split = TimeSeriesSplit(n_splits=3)
sfs = SequentialFeatureSelector(model, n_features_to_select=15, direction='forward', cv=split)
```

```python
removed_columns = ['Player ID', 'Date', 'Name', 'Tm', 'Opp', 'target', 'Season' ]
```

```python
selected_columns = predictor_df.columns[~predictor_df.columns.isin(removed_columns)]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
predictor_df[selected_columns] = scaler.fit_transform(predictor_df[selected_columns])
```

```python
sfs.fit(predictor_df[selected_columns], predictor_df['target'])
```

```python
predictors = list(selected_columns[sfs.get_support()])
```

```python
predictors
```

```python
#We can not do typical cross validation because we have time series data, so we need to right
#a function to ensure that training data is from chronologically earlier than test data
def backtest(data, model, predictors, start=4, step=1):
    all_predictions = [] #List of DF
    seasons = sorted(data['Season'].unique())
    
    for i in range(start, len(seasons), step):
        season = seasons[i]
        
        train = data[data['Season'] < season]
        test = data[data['Season'] == season]
        
        model.fit(train[predictors], train['target'])
        
        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        
        combined = pd.concat([test['target'], preds], axis=1)
        combined.columns = ['actual', 'prediction']
        
        all_predictions.append(combined)
    
    return pd.concat(all_predictions)
```

```python
predictions = backtest(predictor_df, model, predictors)
```

```python
predictions
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

predictions = predictions[predictions['actual'] != -1]

mae = mean_absolute_error(predictions['actual'], predictions['prediction'])
mse = mean_squared_error(predictions['actual'], predictions['prediction'])
rmse = mean_squared_error(predictions['actual'], predictions['prediction'], squared=False)
r2 = r2_score(predictions['actual'], predictions['prediction'])

print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R-squared: {r2}")
```

```python
def find_player_averages(player):
    rolling = player[selected_columns].rolling(10).mean()
    non_numeric_columns = player[['Player ID', 'Season']]
    result = pd.concat([non_numeric_columns, rolling], axis=1)
    return result

predictor_df_rolling = predictor_df_rolling.groupby(['Player ID', 'Season'], group_keys=False).apply(find_player_averages)
```

```python
rolling_columns = [f"{col}_10" for col in predictor_df_rolling.columns]
predictor_df_rolling.columns = rolling_columns

predictor_df = pd.concat([predictor_df, predictor_df_rolling], axis=1)
predictor_df.dropna(inplace=True)
predictor_df
```

```python
removed_columns = list(predictor_df.columns[predictor_df.dtypes == 'object']) + removed_columns + ['Season_10']
removed_columns
```

```python
selected_columns = predictor_df.columns[~predictor_df.columns.isin(removed_columns)]
```

```python
sfs = SequentialFeatureSelector(model, n_features_to_select=30, direction='backward', cv=split)
sfs.fit(predictor_df[selected_columns], predictor_df['target'])
```

```python
predictors = list(selected_columns[sfs.get_support()])
predictors
```
